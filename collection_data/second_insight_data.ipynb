```python
import yfinance as yf
import pandas as pd
```

```python
tickerslist = ['MSFT', 'AAPL', 'NVDA', 'GOOG', '2222.SR', 'AMZN', 'META', 'BRK-B', 'TSLA', 'V', 'MA', 'COST', 'CRM', 'BAC', 'KO']
tickers = yf.Tickers("MSFT AAPL NVDA GOOG 2222.SR AMZN META BRK-B TSLA V MA COST CRM BAC KO")
```

```python
hist = tickers.history(period="5y")
hist.to_csv('stock-history', encoding='utf-8')
```

```python
hist
```

```python
dfs = []
tickers.tickers
```

```python
for ticker in tickerslist:
    t = yf.Ticker(ticker)
    data = t.quarterly_cashflow
    if data is not None:  # Some tickers may not have quarterly cash flow data
        data['Ticker'] = ticker  # Add a column for the ticker symbol
        dfs.append(data)

        # Fetch earnings dates
        earnings_dates = t.get_earnings_dates(limit=16)
        if earnings_dates is not None:
            earnings_df = pd.DataFrame(earnings_dates, columns=['Earnings Dates'])
            earnings_df['Ticker'] = ticker
            dfs.append(earnings_df)
```

```python
combined_df = pd.concat(dfs)

combined_df.reset_index(inplace=True)
```

```python
combined_df
```

```python
combined_df.to_csv('combined-quarterly', encoding='utf-8')
```
